```python
x1 = ['2023년 01월']
y1 = df.iloc[0,:3].sum() # df['0~14세'] 
y2 = df.iloc[0,3:13].sum() # df['15~64세']
y3 = df.iloc[0,13:].sum() # df['65세이상']
```

```python
plt.bar(x1,y1,label='0~14세')
plt.bar(x1,y2,bottom=y1,label='15~64세')
plt.bar(x1,y3,bottom=y1+y2,label='65세 이상')
for i,txt in enumerate(y1):
    plt.text(x1[i],y1[i],txt,ha='center')
for i,txt in enumerate(y2):
    plt.text(x1[i],y1[i]+y2[i],txt,ha='center')
for i,txt in enumerate(y3):
    plt.text(x1[i],y1[i]+y2[i]+y3[i],txt,ha='center')
plt.legend(loc='upper center',nrows=1)
plt.show()
```
