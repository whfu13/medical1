```python
import pandas as pd
df = pd.read_csv('../z20_data/chipotle.tsv', sep='\t')
df
```

```python
# chipotle.tsv에서 주문중 가장 많이 주문한 메뉴 10개를 꺾은선 그래프로 출력하시오.
# 주문별 그룹
# quantity의 합을 구해야 함.
# chipo = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:10]
```

```python
# item_name 별 주문한 개수의 합 출력하시오.
# item_name별 주문한 횟수가 몇번인지 출력하시오.
df['item_name'].value_counts()[:5]
chicken = df['item_name'].value_counts()[:5].values
chipo = df.groupby('item_name')['quantity'].sum().sort_values(ascending=False)[:5]
# df.groupby('item_name')['quantity'].sum()[:5].sort_values(ascending=False).index
# df.groupby('item_name')['quantity'].sum()[:5].sort_values(ascending=False).values
```

```python
x = chipo.index
y = chipo.values
y2 = chicken

plt.figure(dpi=200)
plt.title('가장 많이 주문한 메뉴')
plt.plot(x,y,label='주문수')
plt.plot(x,y2,label='주문횟수')
plt.xticks(rotation=10)
plt.legend(loc='upper center')
```
