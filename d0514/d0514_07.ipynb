```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
# matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)[['이름','국어','영어','수학']]
```

```python
x = df.head(3)['이름']
y = df.head(3)['국어']
y2 = df.head(3)['영어']
y3 = df.head(3)['수학']
```

```python
plt.plot(x,y,label='국어',marker='D')
plt.plot(x,y2,label='영어',marker='o',ls='--')
plt.plot(x,y3,label='수학',marker='s',ls='-.')
for i,txt in enumerate(y3):
    plt.text(x[i],y3[i]+1.0,txt,ha='center')
# ncol: 범례 1줄에 3개까지
plt.legend(loc='upper center',ncol=3)
plt.show()
```

```python
colors = ['r','g','b','y','purple']
plt.bar(x,y,label='국어',color=colors)
# plt.plot(x,y2,label='영어',marker='o',ls='--')
# plt.plot(x,y3,label='수학',marker='s',ls='-.')
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+1.0,txt,ha='center')

# plt.xticks(rocation=90) # x축 각도 변경
# ncol: 범례 1줄에 3개까지
plt.legend(loc='upper right')
plt.show()
```

```python
bar = plt.bar(x,y,label='국어',width=0.2,alpha=0.5)
# 패턴넣기
bar[0].set_hatch('/')
bar[1].set_hatch('x')
bar[2].set_hatch('..')
# plt.plot(x,y2,label='영어',marker='o',ls='--')
# plt.plot(x,y3,label='수학',marker='s',ls='-.')
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+1.0,txt,ha='center')
    

# ncol: 범례 1줄에 3개까지
plt.legend(loc='upper right')
plt.show()
```
