```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df
```

#### fillna - Nan데이터 처리함수

```python
df.fillna('')
df.fillna('없음')
```

```python
df['학교'] = "디지털고"
import numpy as np
df['학교'] = np.nan
```

```python
# 데이터 전체에 Nan 부분을 처리
df.fillna('없음',inplace=True)
df
```

```python
# 해당 컬럼에서 Nan 부분을 처리
df['SW특기'].fillna('없음',inplace=True)
df
```

```python
df
```

#### dropna - Nan처리 : 지우기 
axis : index,colums \
how : any,all

```python
# Nan데이터가 들어가 있는 컬럼을 전체 삭제
df.dropna(axis='columns',inplace=True)
```

```python
df
```

```python
df['SW특기'].dropna(inplace=True)
```

```python
# Nan데이터가 들어가 있는 컬럼에서 모든 컬럼이 Nan인것만 삭제
df.dropna(axis='columns',how='all',inplace=True)
```

```python
# Nan데이터가 들어가 있는 컬럼에서 1개라도 컬럼이 Nan인것만 삭제
df.dropna(axis='columns',how='any',inplace=True)
```

```python
df.loc['4번','영어'] = np.nan       # 1개만 Nan데이터 처리
```

```python
df.dropna(axis='columns',how='any',inplace=True)
```

```python
df
```

```python
# Nan 들어가 있는 index에서 1개라도 Nan이 포함되어 있으면 삭제
df.dropna(axis='index',how='any',inplace=True)
```
