```python
import pandas as pd
filepath = '../z20_data/chipotle.tsv'
df = pd.read_csv(filepath,sep='\t')
df
```

```python
# 데이터 정보 간략하게 확인
df.head()
```

```python
# 컬럼의 타입 변경
df['order_id'] = df['order_id'].astype(str)
```

```python
# 숫자인 경우 대략적인 통계정보를 보여줌.
df.describe()
```

```python
# 달러 $ 부분을 삭제하시오.
df['item_price'] = df['item_price'].str.replace("$","").astype(float)
df.info()
```

```python
df['item_price'].mean()
```

```python
df['item_name'].unique()
len(df['item_name'].unique())
```

```python
# df['item_name'].count() # 전체 개수 확인
# item_name 개별 개수 확인
df['item_name'].value_counts()
```

```python
item_count = df['item_name'].value_counts()
# zip()과 같은 형태로, value_count에 2개 컬럼을 1개 묶어 list저장
list(item_count.items())
```
