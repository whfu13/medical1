#### 데이터 수정: columns 수정

```python
# 학교 컬럼 모두 디지털고로 변경
df['학교'] = '디지털고'
df
```

```python
# 학교컬럼중에 구로고를 디지털고로 수정
df['학교'].replace({'구로고':'디지털고'},inplace=True)
df
```

```python
## SW특기 컬럼에서 java가 포함되어 있는 열을 출력하시오.
flit = df['SW특기'].str.lower().str.contains('java',na=False)
df[flit]
```

```python
df['SW특기'] = df['SW특기'].str.lower()
df
```

```python
df.fillna("none",inplace=True)
df
```

```python
# 학교 문구를 수정
df['학교'] = df['학교'] + '등학교'
df
```

```python
# 과학점수를 +5점을 더해서 출력하시오.
df['과학'] = df['과학'] +5
df
```

```python
# 국어점수를 +10점을 더해서 출력하시오.
df['국어'] = df['국어'] + 10
df
```
