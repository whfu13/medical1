### x,y축 설명 

```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False

x = ['han','kim','hong']
y = [20,40,80]
y2 = [50,30,90]
y3 = [80,90,95]

# 범례 이름지정
plt.plot(x,y,label='국어')
plt.plot(x,y2,label='영어')
plt.plot(x,y3,label='수학')

# 범례 표시, 위치설정, ncols: 1줄에 범례를 3개씩 출력
plt.legend(loc=4,ncols=3)

# 상단타이틀 글자변경, 크기변경
plt.title('막대그래프',fontdict={'family':'Malgun Gothic','size':30})

# x,y축의 설명문구, 위치, 색상
plt.xlabel('국어',loc='left',color='aqua') # left,center,right
plt.ylabel('이름',loc='top',color='green') # top,center,bottom
```
