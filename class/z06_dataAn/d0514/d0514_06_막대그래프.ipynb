```python
import matplotlib.pyplot as plt
import matplotlib

matplotlib.rcParams['font.family'] = 'Malgun Gothic' # 한글설정
# matplotlib.rcParams['font.size'] = 15 # 글자크기
matplotlib.rcParams['axes.unicode_minus'] = False
```

```python
import pandas as pd
df = pd.read_excel('../z20_data/score.xlsx',index_col='지원번호')
df.head(3)
```

```python
# 합계 컬럼을 생성해서 막대그래프, 옆으로 누운 그래프를 출력하시오.
# 합계 점수를 출력
df['합계'] = df['국어']+df['영어']+df['수학']+df['과학']+df['사회']
```

```python
df
```

```python
x = df['이름']
y = df['합계']


plt.bar(x,y,label='합계',color='green')
plt.ylim(0,500)
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+2.0,txt,color='blue')
plt.legend(loc='upper center')
plt.show()
```

```python
x = df['이름']
y = df['합계']


plt.barh(x,y,label='합계')
plt.xlim(0,500)
for i,txt in enumerate(y):
    plt.text(y[i]+1.5,x[i],txt,color='blue')
plt.legend(loc='upper center')
plt.show()
```
