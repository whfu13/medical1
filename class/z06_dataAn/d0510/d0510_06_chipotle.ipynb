```python
import pandas as pd
filepath = '../z20_data/chipotle.tsv'
df = pd.read_csv(filepath,sep='\t')
df
```

```python
df['item_name'].value_counts()[0]
df['item_name'].value_counts().values[0]
```

```python
# 한 주문에 10달러 이상 사용한 주문의 id를 출력하시오.
order_count = df.groupby('order_id').sum()
df.groupby('order_id').sum()['item_price']>=20

# row에 해당하는 컬럼을 출력할때 loc 입력
order_count.loc[df.groupby('order_id').sum()['item_price']>=20,['item_price']]
```

```python
# $표시 제거 및 float 타입 변경
df['item_price'] = df['item_price'].str.replace("$","").astype(float)
```

```python
# 주문당 평균 계산금액 출력하시오.
df.groupby('order_id')['item_price'].sum()
```

```python
# 총 주문 수량 : 1832
# df['order_id'].unique()
# 메뉴별 주문 수량을 출력하시오.
# unique() - 중복제거
# value_counts: 컬럼 데이터별 개수를 출력
df['item_name'].value_counts()[:10]
list(df['item_name'].value_counts()[:10].items())

# items() 항목과 개수가 튜플 형태로 구성
item_count = df['item_name'].value_counts()
for i,(val,cnt) in enumerate(item_count.items()):
    print('Top',(i+1),":",val,",",cnt) 
# sort_values: 컬럼 데이터 정렬
# df.groupby('item_name').count().sort_values('order_id',ascending=False)['order_id'][:10]
```

```python
# item당 주문 총량 출력
len(df.groupby('item_name'))
df.groupby('item_name')['quantity'].sum().tail(30)
```

```python
# 각 아이템별 가격을 구하시오.
# 아이템별 주문 개수가 1개인것의 가격을 출력하시오.
df_one_quantity = df[df['quantity']==1]

# 50개별 최저 가격 - 아이템별 1개 주문금액
df_item_price = df_one_quantity.groupby('item_name').min().sort_values('item_price')

# 메뉴별 1개 주문금액
df_item_price['item_price']

# 메뉴개수 : 50
len(df_one_quantity.groupby('item_name').min().sort_values('item_price')['item_price'])
```

```python
# 주문별, (item_price)가 가장 높은 가격순으로 주문개수를 출력하시오.
# 1. 주문별로 그룹화 한 후 가격 역순으로 정렬하시오.
# 주문개수
# df_item_price = df_one_quantity.groupby('item_name').max().sort_values('item_price',ascending=False)
# df_item_price['item_price']
items = df.groupby('order_id').sum().sort_values('item_price',ascending=False)
items[['quantity','item_price']]
```

```python
# 주문당 개수가 가장 많은 상위 5개를 출력하시오.
items1 = df.groupby('order_id').sum().sort_values('quantity',ascending=False)
items1[['quantity','item_price']][:5]
# items1.head()
```
