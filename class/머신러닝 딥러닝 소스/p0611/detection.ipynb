```python
import seaborn as sns
import pandas as pd
import numpy as np

# 이미지 불러오기
import os
import matplotlib.pyplot as plt
# 한글설정
import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 표시 해결
matplotlib.rcParams['font.family'] = 'Malgun Gothic' # windows 사용자
# matplotlib.rcParams['font.family'] = 'AppleGothic Gothic' # Mac 사용자
matplotlib.rcParams['font.size'] = '15' # 글자크기
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical
import cv2

# data 경로 설정
data = '../z20_data/Spanner.v1i.createml/train/'
```

```python
img_name = os.listdir(data)[0]
img_path = os.path.join(data, img_name)

img = cv2.imread(img_path, cv2.IMREAD_COLOR)

start_point = (150, 150) 
end_point = (250, 250) 

color = (27, 111, 245) 
thickness = 10

image = cv2.line(img, start_point, end_point, color, thickness) 

image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

plt.imshow(image_rgb)
plt.title('Image')
plt.axis('off')
plt.show()
```

```python
def resize_image(img,size=(128,128)):
    return cv2.resize(img,size)
resized_img = resize_image(img)

imageRectangle = img.copy()
start_point = (257,172) 
end_point = (235,250)
color = (237, 7, 49)
thickness = 2
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
cv2.rectangle(imageRectangle, start_point, end_point, color, thickness)
plt.imshow(cv2.cvtColor(imageRectangle, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
cv2.rectangle(imageRectangle, start_point, end_point, color, thickness)
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.title('Resized Image')
plt.axis('off')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

def he_img(img):
    # Ensure the image is in grayscale
    gray_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    return cv2.equalizeHist(gray_img)

# Assume img is already defined and is a BGR image
equalized_img = he_img(img)

imageRectangle = img.copy()
start_point = (257,172) 
end_point = (235,250)
color = (237, 7, 49)
thickness = 2

plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
cv2.rectangle(imageRectangle, start_point, end_point, color, thickness)
plt.imshow(cv2.cvtColor(imageRectangle, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

# Equalized Image
plt.subplot(1, 2, 2)
plt.imshow(cv2.cvtColor(equalized_img, cv2.COLOR_BGR2RGB))
plt.title('Equalized Image')
plt.axis('off')
plt.show()
```

```python
def clahe_image(img):
    clahe = cv2.createCLAHE(clipLimit=2.,tileGridSize=(8,8))
    cl_img = clahe.apply(img)
    return cl_img

imageRectangle = img.copy()
start_point = (257,172) 
end_point = (235,250)
color = (237, 7, 49)
thickness = 2
plt.figure(figsize=(15, 10))
plt.subplot(1, 2, 1)
cv2.rectangle(imageRectangle, start_point, end_point, color, thickness)
plt.imshow(cv2.cvtColor(imageRectangle, cv2.COLOR_BGR2RGB))
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('CLAHE Image')
plt.imshow(clahe_image(img), cmap='gray')
plt.axis('off')
plt.show()
```
