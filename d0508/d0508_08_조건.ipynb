```python
import pandas as pd
df = pd.read_excel('score.xlsx',index_col='지원번호')
df
```

### 데이터 선택 - 조건

```python
# a_list = [1,3,5,7]
a_list = [i*2 for i in range(0,4)] # 리스트의 번호는 lioc
a_list
df.iloc[a_list]
```

```python
# pandas 조건문(1) - 컬럼의 조건을 부여해서 출력시킴
filt = df['키']>190
df[filt]
```

```python
# pandas 조건문(2)
df[df['키']>190]
```

```python
## 키가 평균보다 큰 학생을 출력하시오.
filt = df['키'] >= df['키'].mean()
df.loc[filt,'키']
df.loc[~filt,'키'] # ~: 조건에 반대
```

```python
## 국어 85점 이상, 영어 85점이상 이름,학교,국어,영어 출력하시오.
## and: &, or:|
score = (df['국어'] >= 85) & (df['영어'] >= 85)
df.loc[score,['이름','학교','국어','영어']]
```
