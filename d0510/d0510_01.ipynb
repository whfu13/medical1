```python
## 총합 컬럼을 만들어서, 국어,영어,수학,과학,사회 합계 점수를
#  총합에 입력해서 출력하시오.
```

```python
df['총합'] = df['국어'] + df['영어'] + df['수학'] + df['과학'] + df['사회']
df['총합'].sum()
```

```python
df.fillna(0,inplace=True)
```

```python
df['국어'] = df['국어'].astype(int)
```

```python
df.info()
```

```python
df['결과'] = '불합격'
df
```

```python
# 총합 300점 이상이면 합격, 불합격
df['총합'] >= 300
df.loc[df['총합'] >= 300,'결과'] = '합격'

# 400점 이상인 학생은 우수
df.loc[df['총합'] >= 400,'결과'] = '우수'
df
```

```python
# 총합, 결과를 사회 뒤쪽에 옮겨서 출력하시오.
cols = list(df.columns)
df = df[cols[0:8]+cols[9:]+[cols[8]]]
df
```
