```python
df.rename(index={'출생아\xa0수':'출생아 수','합계\xa0출산율':'합계 출산율'},inplace=True)
```

```python
df.info()
```

```python
# 가로 세로 바꾸는법
df = df.T
```

```python
df.loc['합계 출산율']
```

```python
# 엑셀파일을 import할때 index의 보이지 않는 특수문자가 있는지 체크
df.loc['출생아 수']
```

```python
x = df.index
y = df['출생아 수']
x1 = df.index
y1 = df['합계 출산율']
```

```python
plt.figure(figsize=(13,5))
plt.plot(x1,y1,color='orange',marker='o')
for i,txt in enumerate(y1):
    plt.text(x1[i],y1[i]+0.02,txt,ha='center')
plt.ylim(0,1.5)
plt.yticks([0,1])

# plt.plot(df.columns,df.loc['합계 출산율'])
plt.bar(x,y,color='#f09605')
plt.ylabel('출생아 수(천 명)')
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+3.5,txt,ha='center')
plt.ylim(250,700)
plt.yticks([300,400,500,600])
plt.show()
```

```python
plt.plot(x1,y1,color='gold',marker='o')
for i,txt in enumerate(y1):
    plt.text(x1[i],y1[i]+0.02,txt,ha='center')
plt.ylabel('합계 출산율(가임여성 1명당 명)')
plt.ylim(0,1.5)
plt.yticks([0,1])
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,7))
# 상단 타이틀 출력
fig.suptitle('출생아 수 및 합계 출산율')

ax1.bar(x,y,color='#f09605')
ax1.set_ylabel('출생아 수')
ax1.set_ylim(250,700)
ax1.set_yticks([300,400,500,600])
for i,txt in enumerate(y):
    plt.text(x[i],y[i]+3.5,txt,ha='center')


# 1개의 그래프에 내용을 2가지 넣기
ax2 = ax1.twinx()

# 2번째 꺾은선 그래프
ax2.plot(x1,y1,color='gold',marker='o',ms=13,markeredgecolor='w',markeredgewidth=3,lw=5)
ax2.set_ylabel('합계 출산율(가임여성 1명당 명)')
ax2.set_ylim(0,1.5)
ax2.set_yticks([0,1])
for i,txt in enumerate(y1):
    plt.text(x1[i],y1[i]+0.05,txt,ha='center')
# plt.bar(x,y,color='#f09605')
# plt.plot(x1,y1,color='gold',marker='o')
```
